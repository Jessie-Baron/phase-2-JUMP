```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.get_dataset_names()
```

```python
car_crashes = sns.load_dataset("car_crashes")
car_crashes
```

```python
# checking the distribution and descriptive stats
sns.displot(car_crashes, x="total", kind="kde")

plt.yticks([])
plt.ylabel("")

mean = car_crashes.total.mean()
median = car_crashes.total.median()
mode = car_crashes.total.mode()[0]

plt.axvline(x = mean, color="red")
plt.axvline(x= median, color="green")
plt.axvline(x= mode, color="blue")

plt.legend(['Mode', 'Median', "Mean"])

plt.show()
```

```python
# curve seems fairly normal now checking boxplot
car_crashes.groupby("total").agg(["max", "min", "mean", "median"])
```

```python
sns.boxplot(x=car_crashes["total"])
plt.show()
```
